```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
stats = pd.read_csv("/Users/paramjaswal/Desktop/NBa/Player_mvp_stats.csv")

# Set style
sns.set_style("darkgrid")
```

```python
#Histogram of Player's Points
plt.figure(figsize=(10, 6))
sns.histplot(stats['PTS'], bins=30, kde=True)
plt.title('Distribution of Player Points')
plt.xlabel('Points')
plt.ylabel('Number of Players')
plt.show()
```

```python
#Scatter plot of Points vs Assists
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PTS', y='AST', data=stats, alpha=0.6)
plt.title('Points vs Assists')
plt.xlabel('Points')
plt.ylabel('Assists')
plt.show()
```

```python
#Bar plot of Average Points Scored by Teams
plt.figure(figsize=(14, 7))
sns.barplot(x='Team', y='PS/G', data=stats)
plt.title('Average Points Scored by Teams')
plt.xlabel('Team')
plt.ylabel('Average Points per Game')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot of Points vs MVP Share
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PTS', y='Share', data=stats, alpha=0.6)
plt.title('Points vs MVP Share')
plt.xlabel('Points')
plt.ylabel('MVP Share')
plt.show()
```

```python
# Age vs Performance (Points)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age', y='PTS', data=stats)
plt.title('Player Age vs Points')
plt.xlabel('Age')
plt.ylabel('Points')
plt.xticks(rotation=90)
plt.show()
```

```python
# Average performance metrics by Player Position
plt.figure(figsize=(14, 7))
sns.barplot(x='Pos', y='PTS', data=stats)
plt.title('Average Points by Player Position')
plt.xlabel('Position')
plt.ylabel('Average Points')
plt.show()

```

```python
# Correlation Heatmap
correlation = stats[['PTS', 'AST', 'TRB', 'MP', 'FG%', '3P', '3PA', '2P', '2PA']].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()


```

```python
# MVP Share Over Time (assuming 'Year' column exists)
plt.figure(figsize=(14, 7))
sns.lineplot(x='Year', y='Share', data=stats)
plt.title('MVP Share Over Time')
plt.xlabel('Year')
plt.ylabel('MVP Share')
plt.show()
```

```python
top_players = stats.groupby('Player')['PTS'].sum().sort_values(ascending=False).head(10)
fig, ax = plt.subplots(figsize=(14, 7))
ax.barh(top_players.index, top_players.values, color=sns.color_palette("husl", 10))
ax.set_title('Top 10 Players by Total Points')
ax.set_xlabel('Total Points')
ax.set_ylabel('Player')
ax.xaxis.set_ticks([])
plt.tight_layout()
plt.show()
```

```python
# Player Efficiency: Points vs Field Goal Percentage
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PTS', y='FG%', data=stats, alpha=0.6)
plt.title('Points vs Field Goal Percentage')
plt.xlabel('Points')
plt.ylabel('Field Goal Percentage')
plt.show()
```

```python
# 4. Distribution of Player Positions
position_dist = stats['Pos'].value_counts()
plt.figure(figsize=(12, 7))
position_dist.sort_values().plot(kind='barh', color=sns.color_palette("husl", len(position_dist)))
plt.title('Distribution of Player Positions')
plt.xlabel('Number of Players')
plt.ylabel('Position')
plt.show()
```

```python
# Age Distribution of Players
plt.figure(figsize=(10, 6))
sns.histplot(stats['Age'], bins=30, kde=True)
plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Number of Players')
plt.show()
```

```python
# Average Points over Seasons
plt.figure(figsize=(14, 7))
sns.lineplot(x='Year', y='PTS', data=stats, ci=None, marker='o')
plt.title('Average Points Over Seasons')
plt.xlabel('Season')
plt.ylabel('Average Points')
plt.xticks(rotation=45)
plt.show()
```

```python
# Average Assists over Seasons
plt.figure(figsize=(14, 7))
sns.lineplot(x='Year', y='AST', data=stats, ci=None, marker='o', color='salmon')
plt.title('Average Assists Over Seasons')
plt.xlabel('Season')
plt.ylabel('Average Assists')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap for Win-Loss Ratios Over Seasons
win_loss_data = stats.pivot_table(index='Team', columns='Year', values='W/L%', aggfunc='mean')

plt.figure(figsize=(16, 12))
sns.heatmap(win_loss_data, cmap='coolwarm', cbar_kws={'label': 'Win-Loss Ratio'})
plt.title('Win-Loss Ratios Over Seasons')
plt.xlabel('Season')
plt.ylabel('Team')
plt.show()
```

```python
#Total Points by Team
team_points = stats.groupby('Team')['PTS'].sum().sort_values(ascending=False)
plt.figure(figsize=(14, 7))
team_points.plot(kind='bar', color=sns.color_palette("husl", len(team_points)))
plt.title('Total Points by Team')
plt.xlabel('Team')
plt.ylabel('Total Points')
plt.show()
```

```python
#Average Player Age by Team
team_avg_age = stats.groupby('Team')['Age'].mean().sort_values(ascending=False)
plt.figure(figsize=(14, 7))
team_avg_age.plot(kind='bar', color=sns.color_palette("husl", len(team_avg_age)))
plt.title('Average Player Age by Team')
plt.xlabel('Team')
plt.ylabel('Average Age')
plt.show()
```

```python
#Total MVP Shares by Team
team_mvp_shares = stats.groupby('Team')['Share'].sum().sort_values(ascending=False)
plt.figure(figsize=(14, 7))
team_mvp_shares.plot(kind='bar', color=sns.color_palette("husl", len(team_mvp_shares)))
plt.title('Total MVP Shares by Team')
plt.xlabel('Team')
plt.ylabel('MVP Shares')
plt.show()
```

```python
# Stacked Bar Plot for Team-wise Performance Over Seasons
team_season_points = stats.groupby(['Year', 'Team'])['PTS'].sum().unstack()
plt.figure(figsize=(16, 9))
team_season_points.plot(kind='bar', stacked=True, figsize=(16, 9), colormap='tab20c')
plt.title('Team-wise Performance Over Seasons')
plt.xlabel('Season')
plt.ylabel('Total Points')
plt.legend(title='Teams', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# MVP Shares Across Seasons
plt.figure(figsize=(14, 7))
sns.boxplot(x='Year', y='Share', data=stats, palette='coolwarm')
plt.title('Distribution of MVP Shares Over Seasons')
plt.xlabel('Season')
plt.ylabel('MVP Share')
plt.xticks(rotation=45)
plt.show()
```

```python
# 1. Highest Score in a Single Match Each Season
highest_score_in_match_each_year = stats.groupby('Year')['PS/G'].max()

plt.figure(figsize=(14, 7))
highest_score_in_match_each_year.plot(kind='line', marker='o', color='b')
plt.title('Highest Score in a Single Match Each Season')
plt.xlabel('Season')
plt.ylabel('Points in a Match')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
# 2. Highest Assists in a Single Match Each Season
highest_assists_in_match_each_year = stats.groupby('Year')['AST'].max()

plt.figure(figsize=(14, 7))
highest_assists_in_match_each_year.plot(kind='line', marker='o', color='r')
plt.title('Highest Assists in a Single Match Each Season')
plt.xlabel('Season')
plt.ylabel('Assists in a Match')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
